### 버블정렬
<img src="https://blog.kakaocdn.net/dn/c4wQHp/btrJFPfz5sX/cztROqayVs8n8MRWJLdRX1/img.gif"/><br>
* 인접해있는 요소간 비교를 통해 정렬
* 단순한 알고리즘
* 비효율

```python
def bubble_sort(array):
    for i in range(len(array)):
        for j in range(len(array)-i-1):
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j]
    return array
```

### 삽입정렬
<img src="https://blog.kakaocdn.net/dn/SHuOF/btrJA4MAAQL/odyyoSURJ7MFJlcOsk1Z61/img.gif"/>

* 특정 원소의 작은 index를 가진 원소들을 탐색해서 정렬하는 알고리즘
* 0번째 인덱스 원소는 비교할 작은 인덱스를 가지지 않기 때문에 1번째 인덱스부터 진행함




```python
def insert_sort(array):
    for i in range(1, len(array)):
        for j in range(i, 0, -1):
            if array[j-1] > array[j]:
                array[j-1], array[j] = array[j], array[j-1]
    return array
```

### 선택정렬
<img src="https://blog.kakaocdn.net/dn/bb8tQu/btrJU0B7qfY/WUZh6MNJtJqkdWdKl1ncK0/img.gif"/>

* 배열에서 정렬이 다 될 때까지 최소값을 반복적으로 찾아 정렬하는 방법
* 처리 안 된 가장 작은 데이터를 <b>선택</b> -> 맨 앞 데이터와 바꾸기를 반복


```python
def selection_sort(array):
    for i in range(len(array)):
        min_idx = i
        for j in range(i+1, len(array)):
            if array[min_idx] > array[j]:
                min_idx = j
        array[min_idx], array[i] = array[i], array[min_idx]
    return array
```

### 병합정렬
<img src="https://blog.kakaocdn.net/dn/cGGmSI/btrJ70BiCLE/rqEPArt3AJMpp9u0tKsnL0/img.gif"/>
* 원소가 하나만 남을 때까지 반복적으로 이분할하여 대소비교 후 다시 재배열한 후 원래 크기 형태 배열로 병합
<hr>
ex)  예를 들어 배열 [6, 5, 1, 4, 3, 2, 8, 7]이 있을 때, 첫 번째로 [6, 5, 1, 4]와 [3, 2, 8, 7]로 분리한다. 두 번째로 [6, 5], [1, 4], [3, 2], [8, 7]로 나눈다. 세 번째로 [6], [5], [1], [4], [3], [2], [8], [7]로 나눈다. 이렇게 모든 원소가 분리되면 대소 관계를 고려하여 병합 과정을 거친다. 첫 번째로 [5, 6], [1, 4], [2, 3], [7, 8]이 되며, 두 번째는 [1, 4, 5, 6], [2, 3, 7, 8]이 된다. 마지막으로 하나의 배열로 병합되면서 [1, 2, 3, 4, 5, 6, 7, 8]와 같이 정렬이 완료되면서 알고리즘이 종료된다.

```python
def merge_sort(list):
    if len(list) <= 1:
        return list
    mid = len(list) // 2
    leftList = list[:mid]
    rightList = list[mid:]
    leftList = merge_sort(leftList)
    rightList = merge_sort(rightList)
    return merge(leftList, rightList)
```

```python
def merge(left, right):
    result = []
    while len(left) > 0 or len(right) > 0:
        if len(left) > 0 and len(right) > 0:
            if left[0] <= right[0]:
                result.append(left[0])
                left = left[1:]
            else:
                result.append(right[0])
                right = right[1:]
        elif len(left) > 0:
            result.append(left[0])
            left = left[1:]
        elif len(right) > 0:
            result.append(right[0])
            right = right[1:]
    return result
```

### 파이썬의 정렬
<p style="color:red">기존 리스트 변경</p>

* sort(), sort(reverse=False): 오름차순

```python
li1 = [7,2,9,3,1]
li1.sort()
print(li1)
```

* sort(reverse=True): 내림차순

```python
li1.sort(reverse=True) 
print(li1)
```

```python
li2 = ['c','d','a','b']
li2.sort()
print(li2)
```

cf) reverse(): 자기 자신을 내림차순이 아니라 그냥 원소를 거꾸로 갖게함

```python
li = [9,3,0,7,2,9]
li.reverse()
print(li)
```

<p style="color:red">기존 리스트는 두고 새 리스트 반환</p>

* sorted(): 오름차순 정렬

```python
li3 = [5, 4, 3, 2, 1]
li4 = sorted(li3)
print(li3)
print(li4)
```

* reversed(): 내림차순

```python
li5 = [1, 2, 3, 4, 5]
li6 = reversed(li5)
print(li5)
print(li6)
print(list(li6))
```
