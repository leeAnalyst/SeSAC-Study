# 삼각형의 완성조건

선분 세 개로 삼각형을 만들기 위해서는 다음과 같은 조건을 만족해야 합니다.

가장 긴 변의 길이는 다른 두 변의 길이의 합보다 작아야 합니다.
삼각형의 두 변의 길이가 담긴 배열 sides이 매개변수로 주어집니다. 나머지 한 변이 될 수 있는 정수의 개수를 return하도록 solution 함수를 완성해주세요.

문제 분석
1. a + b > C   C < a +b
2. a + c > b   C > -a + b

```python
sides = [11, 7]
sides = sorted(sides)   # [7, 11]

result = 0
for i in range(sum(sides)):   # 1. a + b > c 조건 충족
    if i > sides[1] - sides[0]:  # 2. b - a < c 조건 충족
        result += 1
print(result)
```

# 숨어있는 숫자의 덧셈 (2)

문제 분석
1. 숫자와 영어를 구분해야한다.
2. 영어는 소문자 대문자가 있다.
3. 붙어있는 숫자를 어떻게 처리하지?

한솔샘의 도움 <br>
v = [] 만들어서 <br>
보다가 문제에 맟닥드리면 v 를 초기화

```python
m = 'aAb1B2cC34oOp526'

m = list(m)

v = 0
result = 0
for word in m:
    if word.isdigit() == True: # isdigit() 숫자인지 문자인지 판별
        if v != 0:
            v += word
        else:
            v = word
    else:
        result += int(v)
        v = 0
result += int(v)
print(result)  
```

# 다항식더하기

다항식 더하기
문제 설명
한 개 이상의 항의 합으로 이루어진 식을 다항식이라고 합니다. 다항식을 계산할 때는 동류항끼리 계산해 정리합니다. 덧셈으로 이루어진 다항식 polynomial이 매개변수로 주어질 때, 동류항끼리 더한 결괏값을 문자열로 return 하도록 solution 함수를 완성해보세요. 같은 식이라면 가장 짧은 수식을 return 합니다.


문제 분석
1. polynomial은 0부터 9까지의 정수, 공백, ‘x’, ‘+'로 이루어져 있습니다.
2. 문자와 숫자 사이의 곱하기는 생략합니다.


```python
polynomial = "1"
p = polynomial.split()  # ['33x', '+', '7', '+', 'x']

# 1. 자연수부분
num = []
for i in p:
    if i.isdigit() == True:
        num.append(int(i))
num = str(sum(num))

#2. x 부분
X = []
for i in list(p):
    if 'x' in list(i):
        X.append(i)  # ['33x', 'x']
for idx in range(len(X)):
    if X[idx] == 'x':
        X[idx] =  '1x'
X2 = []
for i in ''.join(X).split('x'):  # ['33', '1', '']
    if i.isdigit() == True:
        X2.append(int(i))
xx = str(sum(X2)) + 'x'    # '34x'
if xx == '1x':
    xx = 'x'
    
#3. 결합
if num == '0':
    answer = str(xx)
elif xx == '0x':
    answer = str(num)
else:
    answer = str(xx + ' + ' + num)

print(answer)
```
