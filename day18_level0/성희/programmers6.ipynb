# 겹치는 선분의 길이

문제분석
1. y0 > x1 로 겹치는 지 확인 --- X
2. 그려보았더니 range이용 --- 단순히 숫자리스트 넣기 X

1. 겹치는 부분을 체크, 공식 찾기 --- X


```python
lines = [[0, 5], [3, 9], [1, 10]]  # lines의 선상을 점으로 표현

dots = []
for line in lines:
    for dot in range(line[0], line[1]):   # 유의할점이 각 line의 마지막 점은 들어가지 않아야한다.
        dots.append(dot)
print(dots)                             # [0, 1, 2, 3, 4, 3, 4, 5, 6, 7, 8, 1, 2, 3, 4, 5, 6, 7, 8, 9]

dic = {}
for dot in dots:
    dic[dot] = dots.count(dot)          # 겹치는게 많으므로 dictionary 로  숫자를 세어주게한다
print(dic)                              # {0: 1, 1: 2, 2: 2, 3: 3, 4: 3, 5: 2, 6: 2, 7: 2, 8: 2, 9: 1}
    
answer = []
for k, v in dic.items():             
    if v > 1:
        answer.append(k)                # 겹치는 점만 append    
print(answer)
print(len(answer))
```

# 특이한 정렬

문제 설명<br>
정수 n을 기준으로 n과 가까운 수부터 정렬하려고 합니다. 이때 n으로부터의 거리가 같다면 더 큰 수를 앞에 오도록 배치합니다.<br>
정수가 담긴 배열 numlist와 정수 n이 주어질 때 numlist의 원소를 n으로부터 가까운 순서대로 정렬한 배열을 return하도록 solution 함수를 완성해주세요.


문제 분석
1. 정수 n 을 가장먼저 append
2. 가까운수를 abs로 , 그런후 큰수를 앞으로

n-num 이 가장 가까운 것을 앞으로 아닌 것을 뒤로

```python
numlist = [10000,20,36,47,40,6,10,7000]
numlist = sorted(numlist, reverse = True)
n = 30

dic = {}
for num in numlist:
    dic[num] = abs(n-num)  # dic = {10000: 9970, 7000: 6970, 47: 17, 40: 10, 36: 6, 20: 10, 10: 20, 6: 24}
print(dic)                 # dic = {numlist : 거리}

value = []
for v in dic.values():
    if v in value:         # value 값이 겹치는 경우는 append 하지않는다. 왜냐하면 다음 for구문에서 쓰이기 때문
        pass
    else:
        value.append(v)    # value값을 리스트화 시키기
value = sorted(value)      # value = [6, 10, 17, 20, 24, 6970, 9970]
print(value)

result = []
for V in value:            # value = [6, 10, 17, 20, 24, 6970, 9970]
    for k,v in dic.items(): # dic = {10000: 9970, 7000: 6970, 47: 17, 40: 10, 36: 6, 20: 10, 10: 20, 6: 24}
        if V == v:
            result.append(k)  # 예시 V=6 일떄   V==v 이려면  k, v = 36, 6 일때이고 그때 k=36 이 append 된다.
print(result)
```
