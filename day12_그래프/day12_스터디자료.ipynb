# 그래프 (1/27)

* 그래프는 vertex와 edge로 구성된 한정된 자료구조를 의미한다. vertex는 정점, edge는 정점과 정점을 연결하는 간선이다.

(https://ko.wikipedia.org/wiki/%EA%B7%B8%EB%9E%98%ED%94%84_(%EC%9E%90%EB%A3%8C_%EA%B5%AC%EC%A1%B0))



* 구성요소
(https://tipsyziasu.tistory.com/107?category=943494)
    * 노드(Node): 위치
    * 간선(Edge): 위치간의 관계를 표시한 선
    * 인접노드: 서로 이웃인 노드
    
## 무방향 그래프 <br>

* 딱히 정해진 방향이 없어 양방향으로 움직이는 게 가능
    
## 방향 그래프

* 간선을 통해 정해진 방향으로 움직임
    
## 가중치 그래프 <br>

* 간선에 가중치가 할당

* 간선으로 연결되어 있는 노드들은 인접하다(adjacent)라고 표현
    * 인접 행렬(Adjecency matrix): 2차원 배열을 활용하여 그래프를 표현하는 방식
    * 인접 리스트(Adjacent list): 연결 리스트를 활용하여 표현하는 방식
(https://veggie-garden.tistory.com/28_)
